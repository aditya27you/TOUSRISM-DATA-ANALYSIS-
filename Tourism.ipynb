```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
try:

    df = pd.read_csv("Tourism.csv")

except Exception as e:
    print(f"Error reading CSV: {e}")

#  information of the dataset
print("info of the dataset")
print(df.info())
```

```python
# Pehle 5 rows check karo
print("first five rows")
df.head()
```

```python
# Group by season and calculate the average budget required
average_budget_by_season = df.groupby('season')['budget_required_inr'].mean().reset_index()

# Sort by average budget required for better visualization
average_budget_by_season = average_budget_by_season.sort_values('budget_required_inr', ascending=False)

# Plot the average budget required by season
plt.figure(figsize=(12, 6))
sns.barplot(data=average_budget_by_season, x='season', y='budget_required_inr')
plt.title('Average Budget Required by Season')
plt.xlabel('Season')
plt.ylabel('Average Budget Required (INR)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Group by month and calculate the average budget required
average_budget_by_month = df.groupby('month')['budget_required_inr'].mean().reset_index()

# Sort by month for chronological order
average_budget_by_month = average_budget_by_month.sort_values('month')

# Plot the average budget required by month
plt.figure(figsize=(12, 6))
sns.lineplot(data=average_budget_by_month, x='month', y='budget_required_inr')
plt.title('Average Budget Required by Month')
plt.xlabel('Month')
plt.ylabel('Average Budget Required (INR)')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

```python
# Create a datetime column
df['date'] = pd.to_datetime(df['year'].astype(str) + '-' + df['month'].astype(str) + '-01')

# Group by date and calculate the mean visitor growth rate
monthly_growth = df.groupby('date')['visitor_growth_rate'].mean().reset_index()

# Plot the visitor growth rate over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_growth, x='date', y='visitor_growth_rate')
plt.title('Average Visitor Growth Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Average Visitor Growth Rate')
plt.grid(True)
plt.show()
```

```python
# Group by weather and calculate the mean visitor growth rate
growth_by_weather = df.groupby('weather')['visitor_growth_rate'].mean().reset_index()

# Sort by visitor growth rate for better visualization
growth_by_weather = growth_by_weather.sort_values('visitor_growth_rate', ascending=False)

# Plot the visitor growth rate by weather
plt.figure(figsize=(12, 6))
sns.barplot(data=growth_by_weather, x='weather', y='visitor_growth_rate')
plt.title('Average Visitor Growth Rate by Weather')
plt.xlabel('Weather')
plt.ylabel('Average Visitor Growth Rate')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Group by season and calculate the mean visitor growth rate
growth_by_season = df.groupby('season')['visitor_growth_rate'].mean().reset_index()

# Sort by visitor growth rate for better visualization
growth_by_season = growth_by_season.sort_values('visitor_growth_rate', ascending=False)

# Plot the visitor growth rate by season
plt.figure(figsize=(12, 6))
sns.barplot(data=growth_by_season, x='season', y='visitor_growth_rate')
plt.title('Average Visitor Growth Rate by Season')
plt.xlabel('Season')
plt.ylabel('Average Visitor Growth Rate')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Group by event type and calculate the mean visitor growth rate
growth_by_event = df.groupby('event_type')['visitor_growth_rate'].mean().reset_index()

# Sort by visitor growth rate for better visualization
growth_by_event = growth_by_event.sort_values('visitor_growth_rate', ascending=False)

# Plot the visitor growth rate by event type
plt.figure(figsize=(12, 6))
sns.barplot(data=growth_by_event, x='event_type', y='visitor_growth_rate')
plt.title('Average Visitor Growth Rate by Event Type')
plt.xlabel('Event Type')
plt.ylabel('Average Visitor Growth Rate')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Data Analysis Key Findings

*   Numerical columns like 'domestic\_arrivals', 'foreign\_arrivals', 'total\_arrivals', and 'revenue\_inr' are strongly positively correlated.

*   Season significantly influences 'avg\_temp\_c' and 'rainfall\_mm'.
*   'revenue\_inr', 'domestic\_arrivals', and 'foreign\_arrivals' show some variation across seasons, weather conditions, and event types, but these categorical factors may not be the primary drivers on their own.
*   Festivals occur frequently across all seasons, Natural Season events are more concentrated in Monsoon and Autumn, and Wildlife-related events are notably high in Winter.
*   Monsoon season predominantly has Rainy and Cloudy weather, Winter has more Cold and Snowy conditions, and Summer has a mix of Warm and Pleasant weather.
*   The average revenue over time shows fluctuations and seasonal patterns but no clear long-term trend.
